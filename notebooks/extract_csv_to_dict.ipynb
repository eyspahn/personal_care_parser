```python
dupe_inds = df.duplicated(subset=['chemical_name'], keep='first')
```

```python
# build up array of indices we'll remove from the original df.
```

```python
chemicals_with_dupes = []
for index, row in df[dupe_inds].iterrows():
    chemicals_with_dupes.append(row['chemical_name'])
```

```python
chemicals_with_dupes[:5]
```

```python
inds_to_delete = np.array([], dtype='int64')
for chemical_name in chemicals_with_dupes:
    inds_of_df = df[df.chemical_name==chemical_name].sort_values(by=['is_syn', 'datetime_pulled'], ascending=[False, True]).index.values
    inds_to_delete = np.append(inds_to_delete, inds_of_df[1:])
```

```python
inds_to_delete = list(inds_to_delete)
```
